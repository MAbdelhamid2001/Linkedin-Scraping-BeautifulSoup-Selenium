## 1st practice get topics titles

```python
from bs4 import BeautifulSoup
import requests 
from selenium import webdriver
```

```python
from webdriver_manager.chrome import ChromeDriverManager
```

```python
options=webdriver.ChromeOptions()
#options.add_argument('--ignore-certificate-errors')
#options.add_argument('--incognito') #make it in iocoginto 
#options.add_argument('--headless') make the browser window disappear
driver=webdriver.Chrome(ChromeDriverManager().install(),options=options)
```

```python
#in case you don't have any opions use this 
#driver=webdriver.Chrome(ChromeDriverManager().install())
```

```python
url='https://analyticsindiamag.com/'
source=driver.get(url)
```

```python
source_code=driver.page_source
```

```python
soup=BeautifulSoup(source_code,'lxml')

article_block=soup.find_all('h3',attrs={'class':'elementor-post__title'})

article_block
```

```python
d=[]
for titles in article_block:  
    d.append(titles.find('a').get_text().strip())
```

```python
d
```

```python
import pandas as pd
df=pd.DataFrame(data=d)
df
```

## 2nd practice get Quotes

```python
url='https://www.brainyquote.com/topics/motivational-quotes'
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
r=requests.get(url)
soup=BeautifulSoup(r.content,'html5lib')
```

```python
qutos=soup.find_all('div',attrs={'class':'qti-listm'})
```

```python
qutos
```

```python
l=[]
for q in qutos:
    q1=q.img['alt'].split('-')[1]
    q2=q.img['alt'].split('-')[0]
    l.append({'Author':q1,'Quote':q2})
l
```

```python
import pandas as pd
p=pd.DataFrame(l)
```

```python
p
```

# Navigation

```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://pythonprogramming.net/parsememcparseface/'

r=requests.get(url)


soup=BeautifulSoup(r.content,'lxml')
```

```python
nav=soup.nav    
print(nav)
```
